# EMA

```python
import numpy as np
import matplotlib.pyplot as plt

# Serie de tiempo de ejemplo
ts = np.array([1, 2, 4, 5, 4, 3, 5, 7, 9, 8, 7, 6])

# Calcular las EMA con diferentes valores de alpha
alpha_05 = 0.5
alpha_01 = 0.1
alpha_09 = 0.9
ema_05 = np.zeros(len(ts))
ema_01 = np.zeros(len(ts))
ema_09 = np.zeros(len(ts))
ema_05[0] = ts[0]
ema_01[0] = ts[0]
ema_09[0] = ts[0]
for i in range(1, len(ts)):
    print(f'{alpha_05} * {ts[i]} + {(1-alpha_05)} * {ema_05[i-1]} = {alpha_05*ts[i] + (1-alpha_05)*ema_05[i-1]}')
    ema_05[i] = alpha_05*ts[i] + (1-alpha_05)*ema_05[i-1]
    ema_01[i] = alpha_01*ts[i] + (1-alpha_01)*ema_01[i-1]
    ema_09[i] = alpha_09*ts[i] + (1-alpha_09)*ema_09[i-1]

# Graficar la serie de tiempo y las EMA
plt.figure(figsize=(10,5))
plt.plot(ts, label='Serie de tiempo')
plt.plot(ema_05, label='EMA (alpha=0.5)')
plt.plot(ema_01, label='EMA (alpha=0.1)')
plt.plot(ema_09, label='EMA (alpha=0.9)')
plt.legend()
plt.show()
```
