 - Visualizaci√≥n de la trayectoria del gradiente descendiente

```python
import numpy as np
import matplotlib.pyplot as plt

# Funci√≥n y su derivada
def f(x):
    return x**2

def df(x):
    return 2*x

# Learning rate
lr = 0.1

# N√∫mero de iteraciones y el valor inicial de x
n_iter = 10
x = 5

# Lista para almacenar los valores de x y de f(x) en cada iteraci√≥n
x_values = [x]
y_values = [f(x)]

# Se actualiza el valor de x en cada iteraci√≥n
for i in range(n_iter):
    x = x - lr*df(x)
    x_values.append(x)
    y_values.append(f(x))

# Array de valores de x para dibujar la funci√≥n
x_range = np.linspace(-5, 5, 100)
y_range = f(x_range)

plt.plot(x_range, y_range, 'b')
plt.plot(x_values, y_values, 'ro--')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Trayectoria del gradiente descendiente para f(x) = x^2')
plt.show()
```
