```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('../img/astronaut.png', cv2.IMREAD_GRAYSCALE)

# Sobel kernels
sobelx_kernel = np.array([[-1, 0, 1],
                          [-2, 0, 2],
                          [-1, 0, 1]])

sobely_kernel = np.array([[-1, -2, -1],
                          [0, 0, 0],
                          [1, 2, 1]])

# Apply the Sobel kernels
sobelx = cv2.filter2D(image, -1, sobelx_kernel)
sobely = cv2.filter2D(image, -1, sobely_kernel)

# Plot the original image, kernels, and the filtered results
fig, axs = plt.subplots(2, 3, figsize=(12, 8))

axs[0, 0].imshow(sobelx_kernel, cmap='gray')
axs[0, 0].set_title('Sobel X Kernel')
axs[0, 0].axis('off')

axs[0, 1].imshow(sobely_kernel, cmap='gray')
axs[0, 1].set_title('Sobel Y Kernel')
axs[0, 1].axis('off')

axs[0, 2].imshow(image, cmap='gray')
axs[0, 2].set_title('Original Image')
axs[0, 2].axis('off')

axs[1, 0].imshow(sobelx, cmap='gray')
axs[1, 0].set_title('Sobel X Filtered Image')
axs[1, 0].axis('off')

axs[1, 1].imshow(sobely, cmap='gray')
axs[1, 1].set_title('Sobel Y Filtered Image')
axs[1, 1].axis('off')


axs[1, 2].axis('off')

plt.tight_layout()
#plt.savefig('filtered_image.png')
plt.show()
```
